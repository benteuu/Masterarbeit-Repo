```python
import ruptures as rpt
import pandas as pd

def detect_change_points(group, penalty=5):
    """
    Detects change points within a single segment using PELT with 'rbf'.

    Parameters:
    - group: DataFrame containing data for a single segment.
    - penalty: Controls the sensitivity of segmentation.

    Returns:
    - List of detected change points (global indices).
    """
    segment_data = group[['speed', 'acceleration']].values

    if len(segment_data) < 2:  # Skip small segments
        return []

    try:
        algo = rpt.Pelt(model="rbf").fit(segment_data)
        cps = algo.predict(pen=penalty)[:-1]  # Exclude last boundary
        return group.index[cps].tolist()  # Convert to global indices
    except Exception as e:
        print(f"Error processing segment {group['segment'].iloc[0]}: {e}")
        return []

# Apply the function to each segment using groupby()
change_points_dict = unlabelled_df.groupby('segment').apply(detect_change_points)

# Convert to a dictionary (segment → list of change points)
change_points_dict = change_points_dict.to_dict()

print(change_points_dict)  # Output dictionary with segment-wise change points

```

Create new segmnets, where a new segment starts when the trajectory changes, 30mins have passed or a changepoint was detected!

```python
def create_segments_unlabelled(df, change_points_dict):
    """
    Assigns segment numbers based on:
    1. Changes in trajectory.
    2. Changes in label.
    3. Time gaps greater than 30 minutes.
    4. Detected change points.

    Parameters:
    - df: DataFrame with ['trajectory', 'label', 'time'].
    - change_points_dict: Dictionary where keys are segments and values are lists of detected change point indices.

    Returns:
    - df: DataFrame with an updated 'segment' column.
    """
    df = df.sort_values(by=['trajectory', 'time']).copy()  # Ensure correct order

    # Convert time to datetime if needed
    df['time'] = pd.to_datetime(df['time'])

    # Flatten change point indices into a set for quick lookup
    change_points_set = {cp for cps in change_points_dict.values() for cp in cps}

    # Create a boolean mask for segment breaks
    mask = (
        (df['trajectory'] != df['trajectory'].shift()) |  # New trajectory
        (df['label'] != df['label'].shift()) |  # Label change
        (df['time'].diff().dt.total_seconds() > 1800) |  # More than 30 minutes
        (df.index.isin(change_points_set))  # A detected change point
    )

    # Assign segment numbers
    df['segment'] = mask.cumsum()

    return df
```

detect change points over sliding window

```python
def detect_change_points_ruptures(df, window_size=200, penalty=5):
    """
    Detects transportation mode change points using the PELT algorithm in windows.

    Parameters:
    - df: DataFrame with ['speed', 'acceleration']
    - window_size: Number of samples per batch to process (prevents memory issues)
    - penalty: Penalty term (γ) controlling segmentation sensitivity

    Returns:
    - change_points: List of detected change point indices
    """
    change_points = []
    num_samples = len(df)

    for start in range(0, num_samples, window_size):
        end = min(start + window_size, num_samples)
        segment_data = df[['speed', 'acceleration']].iloc[start:end].values

        if len(segment_data) < 2:
            continue  # Avoid errors with very small windows
            #model='mean' for mean shift!
        algo = rpt.Pelt(model="rbf").fit(segment_data)
        cps = algo.predict(pen=penalty)

        # Adjust indices to match full dataset
        adjusted_cps = [start + cp for cp in cps[:-1]]  # Exclude last boundary
        change_points.extend(adjusted_cps)

    return sorted(change_points)

```
