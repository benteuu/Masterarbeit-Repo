```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Example confusion matrix (replace this with your actual data)
cm = np.array([
    [1746, 46, 0, 0, 213],
    [58, 2484, 253, 77, 321],
    [24, 326, 1271, 134, 34],
    [15, 25, 38, 3670, 127],
    [139, 153, 18, 10, 3326]
])

# Normalize by dividing each row by its sum (to get row-wise percentages)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Define labels
labels = ["bike", "bus", "car", "train", "walk"]

# Plot normalized confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized, annot=True, fmt=".2f", cmap="Blues", xticklabels=labels, yticklabels=labels)

# Add titles and labels
plt.title("Normalized Confusion Matrix LSTM Model")
plt.xlabel("Predicted")
plt.ylabel("True")

# Show plot
plt.show()

```
